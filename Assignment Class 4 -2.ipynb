### Data Transformation 

If you look on above data, you will notice there are still some problems. First, like the *Wage* data, shall be float figures, but here you see something like €67M, €420K, so I need to do some data transformation to turn they to float numbers which can be used by the machine learning  model; Second, there are some data need to convert to desired data types, for example, "date" typed data like the *Joined* date of players needs to be transitted to integer numbers. 

```python
# to remove the money value currency $ and M/K, change the scale to M to shrink the cost value
def value_tran(val):
    rt=0
    if val.endswith('M'):
        rt = float(val[1:-1])*1000000
    elif val.endswith('K'):
        rt = float(val[1:-1])*1000
    else:
        rt = float(val[1:])
    return rt/10000000.0
```

```python
# to remove the money value currency $ and M/K, change the scale to M to shrink the cost value
def wage_tran(val):
    rt=0
    if val.endswith('M'):
        rt = float(val[1:-1])*1000000
    elif val.endswith('K'):
        rt = float(val[1:-1])*1000
    else:
        rt = float(val[1:])
    return rt/10000.0
```

```python
# to remove the '
def height_tran(val):
    idx= val.find(r"'")
    rt=0
    
    if idx>0:
        s=val[:idx]+"."+val[(idx+1):]
        rt = float(s)
    else:
        rt = float(val)
    return rt
```

```python
# to remove the lbs
def weight_tran(val):
    rt=0
    if val.endswith('lbs'):
        rt = float(val[:-3])
    else:
        rt = float(val)
    return rt
```

```python
df['Value'] = df['Value'].apply(value_tran)
df['Wage'] =  df['Wage'].apply(wage_tran)
df['Height']= df['Height'].apply(height_tran)
df['Weight']= df['Weight'].apply(weight_tran)
```

You can stop here to see what the "Wage" looks like after the data transformation, first all the currency flag removed , it is pure float data now. And then, the float number is scale to a proper level: from 0.1 to 56. This is important, as the *"Wage"* value is the training target of machine learning model, it will attend the computation, we can't make it too big or else there is additional burden of computing, and the loss value looks very big, hard to make judgement if the model is trained well.

### Nerual network  model through Keras

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
#some learners constantly reported 502 errors in Watson Studio. 
#This is due to the limited resources in the free tier and the heavy resource consumption of Keras.
#This is a workaround to limit resource consumption
import keras
from keras import backend as K
K.set_session(K.tf.Session(config=K.tf.ConfigProto(intra_op_parallelism_threads=1, inter_op_parallelism_threads=1)))
```

Now I will use Keras to build the neural network to do the prediction again. But there is one issue I need to solve first, Keras doesn't offer built-in function of R2(R-squared), I need to implement such metrics function first. Let's see what is R^2:

$$R^2 = 1- \frac{\sum_{i=1}^n{(y_i-\hat{y_i})^2}}{\sum_{i=1}^n{(y_i-\bar{y_i})^2}}$$

The function "r_square" is defined through backend of Keras API(tensor), and will be set as the metrics function of the NN model for evaluation.



```python
# ooops! there is no R^2 in Keras! Have to implement mine
def r_square(y_true, y_pred):
    SS_res =  K.sum(K.square(y_true - y_pred)) 
    SS_tot = K.sum(K.square(y_true - K.mean(y_true))) 
    return ( 1 - SS_res/(SS_tot) )#+ K.epsilon()
```

The NN model is for regression, so I defined the first layer of the network input dimension equals to the PCA dimension of training data, namely 800, and the 1st output cell unit is 800; the second layer cell unit is 400,and both of first 2 layers leveraged "relu" activation; the output layer cell output is 1 and activation is "linear". For this neural network model, I used mean_squared_error as loss function, and pertinent metrics is mse and R^2, and optimizer is "adam". 

The training hyperparameters are epochs=100, batch_size=80 and you can see the fit result is documented here, the loss kept down in cycle, and the R^2 value finally stabilized to close to 1.


```python
def nn_model():
    model = Sequential()
    model.add(Dense(800, input_dim=800, kernel_initializer='normal', activation='relu'))
    model.add(Dense(400, activation='relu'))
    model.add(Dense(1,activation='linear'))
    # Compile model
    model.compile(loss='mean_squared_error',optimizer='adam',metrics=[r_square,"mse"])
    return model
```

```python
model=nn_model()
model.fit(X_train_pca, train_y_w, validation_split=0.2,epochs=100, batch_size=80,verbose=2)
```

At last, let's predict on the X_test test data set, I got the evaluation of performance as :

```python
wage_pred = model.predict(X_test_pca)
```

```python
score = model.evaluate(X_test_pca, test_y_w)
print("Loss:%.3f, R^2:%.3f, MSE:%.3f" % (score[0],score[1],score[2]))
```
